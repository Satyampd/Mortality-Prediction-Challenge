```python
Y = train.iloc[:,-1:]
```

```python
X = train.iloc[:,:-1]
```

```python
for i in X:
    print(i,X[i].dtype)
```

```python
X=X.iloc[:,10:]
```

```python
X = X.drop(['Otnutritdx'],axis=1)
X = X.drop(['Othnervdx'],axis=1)
X =X.drop(['GLUCOSE_MEAN_DAY2'],axis=1)

X.info()
```

```python

count=0
for i in X :
    print(count,i)
    count=count+1
```

```python
for i in X:
    mean_value=X[i].mean()
    X[i]=X[i].fillna(mean_value)
```

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.25, random_state = 42)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=2)
X_train_res, y_train_res = sm.fit_sample(X_Train, np.ravel(Y_Train))
print(sum(y_train_res==0))
print(sum(y_train_res==1))
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train_res = sc_X.fit_transform(X_train_res)
X_Test = sc_X.transform(X_Test)
```

```python

from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 1000, criterion = 'entropy', random_state = 0)
classifier.fit(X_train_res, y_train_res)
```

```python
from sklearn.metrics import accuracy_score
Y_Pred = classifier.predict(X_Test)
accuracy = accuracy_score(Y_Test, Y_Pred)
accuracy
```
